# visualization part

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic-training-data.csv")
```

```python
df.head()
```

```python
df=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```

# univariate(single variable)

```python
sns.countplot(x="Survived",data=df)
plt.show()
```

```python
sns.countplot(x="Sex",data=df)
plt.show()
```

```python
sns.countplot(x="Pclass",data=df)
plt.show()
```

```python
df.describe()
```

```python
sns.boxplot(x="Age",data=df)
plt.show()
```

# first line at 20 is for 25% and next for 50% and for 75

```python
sns.boxplot(x="Fare",data=df)
plt.show()
```

```python
sns.boxplot(x="Parch",data=df)
plt.show()
```

```python
#minimum 25 and 50 and 75 are same
```

```python
sns.boxplot(x="SibSp",data=df)
plt.show()
```

```python
#25 and 50 are 0 and 75 is 1 and max is 8
```

# histograph

```python
plt.hist("Age",data=df)
plt.show()
```

```python
#histograph and boxplot are for numericles
```

# bivariate

```python
sns.countplot(x="Survived",hue="Sex",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Pclass",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Embarked",data=df)
plt.show()
```

```python
# two numericall plot
```

```python
sns.relplot(x="Age",y="Fare",data=df)
```

```python
sns.relplot(x="Fare",y="Pclass",data=df)
```

```python
#one catogorical and one numerical
```

```python
sns.catplot(x="Survived",y="Age",data=df)
```

```python
sns.catplot(x="Survived",y="SibSp",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",data=df,kind="violin")
```
